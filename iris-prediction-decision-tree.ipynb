```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz

# Sources
iris = pd.read_csv('Iris.csv')
tree_model = DecisionTreeClassifier()

# Datasets Information
iris.info()
iris.head()

# Remove ID Column on Datasets
iris.drop('Id', axis=1, inplace=True)

# Separate the Atribute and Label
x = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ]]
y = iris['Species']

# Separate data to Train & Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=123)

# Train model with train data
tree_model = tree_model.fit(x_train, y_train)

# Find the accuracy score to our model
y_pred = tree_model.predict(x_test)
acc_score = round(accuracy_score(y_pred, y_test), 3)
print('\nAccuracy: ', acc_score)

# Input number according to features (x) list
print(tree_model.predict([[6.2, 3.4, 5.4, 2.3]])[0])

# Used to see the visualization result of Decision Tree model that we made
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ],
    rounded= True,
    filled =True)
```
